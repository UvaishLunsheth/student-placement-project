```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
# Load raw dataset
df = pd.read_csv('../data/raw/student_placement_salary_elite_v2.csv')
```

```python
## Validation assumption
df[df["salary_lpa"] == 0]["placed"].value_counts()
```

> #### There are missing values in `company_type` and `job_role` but they did not treated as data quality issues,but as meaningful indicators of unplaced students

```python


# Fill any remaining NaN values
df["company_type"].fillna("Not Placed", inplace=True)
df["job_role"].fillna("Not Placed", inplace=True)
```

```python
df.isnull().sum().sum()
```

```python
df["placement_status"] = df["placed"].map({1: "Placed", 0: "Not Placed"})
```

```python
df.info()
```

```python
## Save cleaned data to processed folder
output_path = '../data/processed/student_placement_cleaned.csv'
df.to_csv(output_path, index=False)
print(f"✓ Cleaned data saved to: {output_path}")
print(f"Shape: {df.shape}")
print(f"Columns: {list(df.columns)}")
```
