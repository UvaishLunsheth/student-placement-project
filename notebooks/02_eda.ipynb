```python
# Exploratory Data Analysis - Student Placement & Salary
## Understanding the Factors Influencing Placement Success and Salary Packages

## Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style and configuration
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['font.size'] = 10
pd.set_option('display.max_columns', None)

## Load Cleaned Data
df = pd.read_csv('../data/processed/student_placement_cleaned.csv')

# Display basic statistics
print("Dataset Shape:", df.shape)
print("\nColumn Names and Types:")
print(df.dtypes)
print("\nFirst few rows:")
df.head()
```

```python

# ==========================================
# PART 1: What drives Placement Success?
# ==========================================

# Select only numeric columns and drop ID + Salary (to prevent data leakage)
numeric_df = df.select_dtypes(include=['number']).drop(columns=['student_id', 'salary_lpa'], errors='ignore')

# Calculate the correlation matrix
corr_matrix = numeric_df.corr()

# Plot the Heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap: Features vs. Placement Status', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

# Print the exact rankings for easy reading
print("--- Top Factors Influencing PLACEMENT ---")
print(corr_matrix['placed'].sort_values(ascending=False))
print("\n")


# ==========================================
# PART 2: What drives High Salaries?
# ==========================================

# Filter for PLACED students only
placed_df = df[df['placed'] == 1].copy()

# Select numeric columns, drop ID and 'placed' (since they are all 1 now)
numeric_placed_df = placed_df.select_dtypes(include=['number']).drop(columns=['student_id', 'placed'], errors='ignore')

# Calculate correlation matrix for salary drivers
salary_corr = numeric_placed_df.corr()

print("--- Top Factors Influencing SALARY (Placed Students Only) ---")
print(salary_corr['salary_lpa'].sort_values(ascending=False))
```

```python
# Scatter plot: Placement vs Backlogs
plt.figure(figsize=(10, 6))

# Create scatter plot with jitter for better visualization
x_jitter = df['backlogs'] + np.random.normal(0, 0.1, len(df))  # Add small random noise
y_jitter = df['placed'] + np.random.normal(0, 0.05, len(df))  # Add small random noise

plt.scatter(x_jitter, y_jitter, alpha=0.6, c=df['placed'], cmap='coolwarm', edgecolor='black', s=50)

# Add trend line
z = np.polyfit(df['backlogs'], df['placed'], 1)
p = np.poly1d(z)
x_trend = np.linspace(df['backlogs'].min(), df['backlogs'].max(), 100)
plt.plot(x_trend, p(x_trend), "r--", linewidth=2, label=f'Trend line')

plt.title('Placement Status vs Number of Backlogs', fontsize=14, fontweight='bold')
plt.xlabel('Number of Backlogs')
plt.ylabel('Placement Status (0 = Not Placed, 1 = Placed)')
plt.yticks([0, 1], ['Not Placed', 'Placed'])
plt.grid(True, alpha=0.3)
plt.legend()

# Add statistics annotation
correlation = df['backlogs'].corr(df['placed'])
plt.annotate(f'Correlation: {correlation:.3f}', xy=(0.05, 0.95), xycoords='axes fraction',
             fontsize=12, bbox=dict(boxstyle="round,pad=0.3", facecolor="white", alpha=0.8))

plt.tight_layout()
plt.show()

# Additional statistics
print("=" * 60)
print("PLACEMENT VS BACKLOGS ANALYSIS")
print("=" * 60)
print(f"Correlation coefficient: {correlation:.3f}")
print(f"Average backlogs (Placed): {df[df['placed'] == 1]['backlogs'].mean():.2f}")
print(f"Average backlogs (Not Placed): {df[df['placed'] == 0]['backlogs'].mean():.2f}")
print(f"Placement rate by backlogs:")
backlog_stats = df.groupby('backlogs')['placed'].agg(['count', 'sum', lambda x: (x.sum()/len(x)*100)]).round(1)
backlog_stats.columns = ['Total Students', 'Placed', 'Placement Rate (%)']
print(backlog_stats)
```

```python
# Scatter plot: Placement vs Backlogs
plt.figure(figsize=(10, 6))

# Create scatter plot with jitter for better visualization
x_jitter = df['resume_score'] + np.random.normal(0, 0.1, len(df))  # Add small random noise
y_jitter = df['placed'] + np.random.normal(0, 0.05, len(df))  # Add small random noise

plt.scatter(x_jitter, y_jitter, alpha=0.6, c=df['placed'], cmap='coolwarm', edgecolor='black', s=50)

# Add trend line
z = np.polyfit(df['resume_score'], df['placed'], 1)
p = np.poly1d(z)
x_trend = np.linspace(df['resume_score'].min(), df['resume_score'].max(), 100)
plt.plot(x_trend, p(x_trend), "r--", linewidth=2, label=f'Trend line')

plt.title('Placement Status vs Resume Score', fontsize=14, fontweight='bold')
plt.xlabel('Resume Score')
plt.ylabel('Placement Status (0 = Not Placed, 1 = Placed)')
plt.yticks([0, 1], ['Not Placed', 'Placed'])
plt.grid(True, alpha=0.3)
plt.legend()

# Add statistics annotation
correlation = df['resume_score'].corr(df['placed'])
plt.annotate(f'Correlation: {correlation:.3f}', xy=(0.05, 0.95), xycoords='axes fraction',
             fontsize=12, bbox=dict(boxstyle="round,pad=0.3", facecolor="white", alpha=0.8))

plt.tight_layout()
plt.show()

# Additional statistics
print("=" * 60)
print("PLACEMENT VS RESUME SCORE ANALYSIS")
print("=" * 60)
print(f"Correlation coefficient: {correlation:.3f}")
print(f"Average resume score (Placed): {df[df['placed'] == 1]['resume_score'].mean():.2f}")
print(f"Average resume score (Not Placed): {df[df['placed'] == 0]['resume_score'].mean():.2f}")
print(f"Placement rate by resume score:")
resume_stats = df.groupby('resume_score')['placed'].agg(['count', 'sum', lambda x: (x.sum()/len(x)*100)]).round(1)
resume_stats.columns = ['Total Students', 'Placed', 'Placement Rate (%)']
print(resume_stats)
```

### "Skill vs Academics" Matrix

```python
print("=== 2. Placement Rates by Segment (Skill vs CGPA) ===")

# We use the median as the cutoff for "High" vs "Low"
cgpa_median = df['cgpa'].median()
skill_median = df['skill_score'].median()

def create_segment(row):
    if row['skill_score'] > skill_median and row['cgpa'] > cgpa_median:
        return '1. High Skill + High CGPA'
    elif row['skill_score'] > skill_median and row['cgpa'] <= cgpa_median:
        return '2. High Skill + Low CGPA'
    elif row['skill_score'] <= skill_median and row['cgpa'] > cgpa_median:
        return '3. Low Skill + High CGPA'
    else:
        return '4. Low Skill + Low CGPA'

# Apply the segmentation
df['student_segment'] = df.apply(create_segment, axis=1)

# Calculate the placement rate (mean of 0s and 1s gives the exact percentage)
segment_analysis = df.groupby('student_segment')['placed'].agg(['mean', 'count']).rename(columns={'mean': 'Placement_Rate', 'count': 'Total_Students'})
segment_analysis['Placement_Rate'] = (segment_analysis['Placement_Rate'] * 100).round(2).astype(str) + '%'
print(segment_analysis)
print("\n" + "="*50 + "\n")
```

```python


# ==========================================
# Finding the Breaking Point: "Below Average" Placements
# ==========================================

# 1. Relaxing thresholds to the "Average/Median" mark (50th percentile)
below_avg_resume = df['resume_score'].median() 
below_avg_coding = df['coding_score'].median() 

# 2. Relaxing categorical/discrete features
low_skill_threshold = 2  # Including average skills (0, 1, and 2)
high_backlogs = 1        # Still keeping at least 1 backlog as a penalty
lower_tiers = [2, 3]     # Now allowing Tier 2 colleges as well

# 3. Filter for students who meet these relaxed "below average" criteria
relaxed_odds_df = df[
    (df['resume_score'] <= below_avg_resume) &
    (df['coding_score'] <= below_avg_coding) &
    (df['skill_score'] <= low_skill_threshold) &
    (df['backlogs'] >= high_backlogs) &
    (df['college_tier'].isin(lower_tiers)) &
    (df['placed'] == 1) 
]

# Output the new thresholds used
print("--- Relaxed Threshold Limits Applied ---")
print(f"Resume Score <= {below_avg_resume:.2f} (Bottom 50%)")
print(f"Coding Score <= {below_avg_coding:.2f} (Bottom 50%)")
print(f"Skill Score <= {low_skill_threshold}")
print(f"Backlogs >= {high_backlogs}")
print(f"College Tier in {lower_tiers}")
print("="*40)

# The New Reveal
print(f"\nTotal 'Below Average' Placements Found: {len(relaxed_odds_df)}")

if len(relaxed_odds_df) > 0:
    print("\n--- Average Salary for these Placements ---")
    print(f"₹{relaxed_odds_df['salary_lpa'].mean():.2f} LPA")
    
    print("\n--- Job Roles They Secured ---")
    print(relaxed_odds_df['job_role'].value_counts())
```



### **1. The "Mediocrity" Threshold is Tight**

Out of the 9,000 students in the dataset, only 110 (~1.2%) managed to get placed despite having below-average resumes, below-average coding scores, low/average general skills, a non-premium college tier, *and* active backlogs.

* **Insight:** The dataset's underlying logic heavily punishes this combination. To recruiters reading your project, this proves your model will have strong, distinct boundaries to learn from.

### **2. The Salary Penalty**

The average salary for these 110 students dropped to **₹59.73 LPA**.

* **Insight:** Remember that the average salary for *all* placed students was around **₹64.75 LPA**. So, even when these below-average students manage to scrape by and get a job, they suffer roughly a ₹5 LPA "penalty" compared to their peers.

### **3. The "Job Role" Synthetic Quirk (Another catch!)**

Look at the distribution of the roles they secured: Data Scientist (35), Software Engineer (27), Analyst (25), Web Developer (23).

* **Insight:** The distribution is almost perfectly even. In the real world, "below average" candidates with backlogs do not predominantly land Data Scientist roles over Web Developer or Analyst roles. This perfectly even spread is the final, definitive proof of how the synthetic data was generated: job roles were likely assigned using a random distribution function rather than being tied to specific skill thresholds.
